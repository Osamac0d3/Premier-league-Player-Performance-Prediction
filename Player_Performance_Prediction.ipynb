```python
import pandas as pd
df=pd.read_csv('premier_league_players_5years_with_rating.csv')
df.sample(2)
```

```python
df=df.drop(columns={'match_number','season' })
```

```python
df.sample(2)
```

```python
df_encoded = pd.get_dummies(df,columns=["player_name","team","opponent_team"],dtype=int)
```

```python
df['player_name'].nunique()
```

```python
df.head(5)
```

```python
df_encoded.head(5)
```

```python
y=df_encoded['rating']
X=df_encoded.drop(columns={'rating'})
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(128, activation="relu", input_shape=(X.shape[1],)))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))

model.add(Dense(1))  # rating prediction

model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

model.fit(X, y, epochs=30, batch_size=64)
```

```python
print(model.predict(X[:2]))
```

```python
y
```

```python
y_predict=model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,y_predict)
mse
```

```python
df_pt = df_pt.groupby(['player_name','team','opponent_team']).mean(numeric_only=True).reset_index()
```

```python
y_pt=model.predict(df_pt_encoded)
y_pt
```

```python
y_pt=pd.DataFrame(y_pt,columns=['predicted_rating'])
y_pt
```

```python
df_actual_pt=pd.concat([df_pt,y_pt],axis=1)
df_actual_pt
```

```python
filtered_df = df_actual_pt[
    (df_actual_pt['team'].isin(['Arsenal','Manchester City'])) &
    (df_actual_pt['opponent_team'].isin(['Arsenal','Manchester City']))
].sort_values(by='predicted_rating', ascending=False)

filtered_df
```
